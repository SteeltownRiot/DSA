<span style="color:#dd7c0d; font-weight:700; font-size:18px">Virtualization</span>

Virtualization is the foundation of cloud computing.
It doesn't create dependencies between the physical resources and your application. 
It builds an abstraction that allows you to program them, but not have to worry about them. 
So, each of the services are going to be based on virtual machines. 
They're going to be based on remote procedure calls because of distribution. 

Virtualization allows you to have distributed computing models mapped across networks onto individual machines. 
And the client server model is used in providing the services to the outside clients of the system.
It will offer simplicity of use, ease of programming, and allow those client server paradigms to be used to construct services. 

The Economics of cloud requires you to share the resources. 
Users see their own piece of the cloud but not other other user's pieces of the cloud. 
There are layers of abstraction that hide this behaviour and users don't realize that they're sharing hardware. 

To understand virtualization, let's build an abstract model of what a generic computing resources should look like. 
We have a background knowledge of what physical computer resources are like. 
Build that abstract model from this physical computing resources. 
We then use this abstract model to do virtualization. 

Let's have a look at Layers of abstraction. 
We go out, we purchase a cloud center. 
We can get lots of hardware. 
We get some servers, some network in the center, some CPUs and storage. 
That physical model then is going to be multiplexed to many different users. 
Virtualization keeps them all independent from tripping on each other.
All this things when put together will be a model of what those systems provide, 
but in an abstract way. 
So those abstractions will get mapped down on to the actual, particular physical entities.

Virtualization is a mechanism to actually map those dependencies onto the real hardware and provide the layer of abstraction that we want.

Containers can help ensure that applications deploy quickly, reliably, and consistently regardless of deployment environment. 

Containers also give you more granular control over resources giving your infrastructure improved efficiency. Running containers in the AWS Cloud allows you to build robust, scalable applications and services by leveraging the benefits of the AWS Cloud such as elasticity, availability, security, and economies of scale. You also pay for only as much resources as you use.

<span style="color:#dd7c0d; font-weight:700; font-size:18px"> SSH Keys</span>

SSH keys serve as a means of identifying yourself to an SSH server using public-key cryptography and [challenge-response authentication](https://en.wikipedia.org/wiki/Challenge%E2%80%93response_authentication). 
One immediate advantage this method has over traditional password authentication is that you can be 
authenticated by the server without ever having to send your password over the network. 
Anyone eavesdropping on your connection will not be able to intercept and crack your password 
because it is never actually transmitted. 
Additionally, using SSH keys for authentication virtually eliminates the risk posed by brute-force
password attacks by drastically reducing the chances of the attacker correctly guessing the proper credentials.

Along with additional security offered, SSH key authentication can be more convenient
than the more traditional password authentication.
When used with a program known as an SSH agent, SSH keys can allow you to connect to a server, 
or multiple servers, without having to remember or enter your password for each system.

The keys are primarily used for automated processes and for implementing passwordless ssh logins 
by system administrators and power users. 
SSH (Secure Shell) is used for managing networks, operating systems, and configurations. 
It is also inside many file transfer tools and configuration management tools.
Every major corporation uses it, in every data center.

The keys enable the automation that makes modern cloud services and other computer-dependent services 
possible and cost-effective. They offer convenience and improved security when properly managed.

<span style="color:#dd7c0d; font-weight:700; font-size:18px"> TYPES OF KEYS IN SSH</span>

SSH uses keys for multiple purposes.
Most common ones are explained below.

<span style="color:#3db220; font-weight:600; font-size:14px; font-style:italic">Authorized Keys - Public User keys:</span>
Authorized keys are public keys that grant access.
They are similar to locks that an identity key (private key) can open.
Authorized keys are configured for an SSH server.
They are typically configured in an authorized_keys file in .ssh subdirectory in the user's home directory.
Identity keys are usually stored in a user's `.ssh` directory, for example, `.ssh/ssh_id_rsa`.

<span style="color:#3db220; font-weight:600; font-size:14px; font-style:italic">Identity Keys - Private User keys:</span> 
Identity keys are private keys that an SSH client uses to authenticate itself when logging into an SSH server.
By default, identity key file names start with id_<algorithm>(in above figure, its id_rsa),
but other names are also often used. 
Identity keys may be password-protected, however the ones used for automated access usually are not. 
In most environments the vast majority of the keys are not protected by a passphrase.
