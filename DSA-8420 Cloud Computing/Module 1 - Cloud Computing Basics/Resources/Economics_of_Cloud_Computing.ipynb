<span style="color:#4286f4; font-weight:700; font-size:28px"> Economics of Cloud Computing</span>

Cloud computing is the trend driving the economy.
Computers are every where and so is the effect of cloud computing in every business that uses computers directly or indirectly.
The economics of cloud computing is called Cloudonomics, the term coined by Joe Weinman.
Looking at the value of utility pricing, yan have cloud services that are more expensive 
than owning your own infrastructure per hour or per day, and still be economical. 

Let's look at an example. Consider a car.
You can buy or lease a car for, let's say \$20 a day.
So if you want to rent a car for a day, it costs you \$50 a day.
So, obviously, renting the car is more expensive than owning the car, per day.
But consider you go on a trip for two days for a vacation.
You don't buy a car to go there. You take an airplane trip and then rent a car for two 
days because buying wouldn't make sense.
So it totally depends on the type of demand that you have to figure out if it's more 
economical to rent or buy a car or services.

Let's do some math here. Following notation is used to illustrate cloud resource usage


* D(t): Demand for resources, $0<t<T$ where resources are needed between time period 0 through T.


* P: max(D(t)): Peak Demand


* A: Avg(D(t)): Average Demand


* B: Baseline(owned) Unit Cost


* $B_T$:Total Baseline Cost


* C: Cloud unit cost


* $C_T$: Total Cloud Cost


* U = C/B: Utility premium

Consider there is demand for compute resources.
Demand is shown here by demands per time.
On a certain day you might have a lot of computing demands, another day your demands could go down.
We are looking at this for a certain time period between 0 and time $T$, let's say $T$ is 365 days.
During this time, demand goes up and down.
So P is the peak demand during a day throughout this 365 days, the maximum demand is denoted by P. 
Average demand, $A$, throughout this time is the mathematical mean of all the demand through out the year. 
B would be the baseline cost of owning computing unit. 

For example, a computer might cost `$10 a day`, that would be $B$ here. 
$B_T$ would be the total cost of owning infrastructure throughout the year. 
Cloud unit cost, $C$ is the cost of cloud resources per day.
Total cost of renting cloud computing resources, throughout the year is $C_T$. 
Finally, the utility premium, is defined as the ratio of cloud cost over baseline cost. 
For example, consider the cloud cost is `$45 and baseline cost is $10`. 
The utility premium would be 4.5. Lets compute the costs.

Even if the demand is erratic throughout the year, and goes up and down you need to provide services to customers every day.
You don't buy new machines every day when there is a peak in demand. 
$P$ is the peak demand implies you need that many machines throughout the year if you want to tackle any kind of traffic load. 
So when the demand is very low, most of the 200 machines will be sitting idle.
This is the reason why cloud computing wins the argument. 

So the total cost of owning, $B_T$ would be the baseline cost, multiplied by the time period,$T$, 
multiplied by how many machines you have, $P$, the maximum number of machines that you need to handle peak demand. 

Lets calculate cloud cost. 
In cloud you can request and release resources from the cloud provider as demand fluctuates.
The cost of the cloud is utility premium, $U$ multiplied by B. 
And you do an integral over time throughout the year. The answer is a lower measure for cloud costs. 

$$C_T = \int_{0}^{T} U * B * D(t).dt = A * U * B * T$$

$$B_T = P * B * T, \quad because\ the\ baseline\ should\ handle\ peak\ demand$$


The question is, when is cloud's cost going to be cheaper than owning? $C_T < B_T$. 

On substituting the values in equation $C_T < B_T$, you get

$$ A * U * B * T < P * B * T  \implies U < P/A$$ 

The math implies that clouds are cheaper than baseline when the utility premium is less than the ratio of peak demand to average demand. 
So if the workload is highly erratic, where average is a certain value and peak is a certain value. 
If the ratio of peak to average is more than the utility premium then it becomes more economical to use cloud computing. 

In real world, demands often flucuate a lot. 
Imagine the traffic on Twitter on world series championship day or on US presidential election day 
or during black friday sale on amazon.
The key factor is to figure out the ratio of your peak demand to average demand. 
In hybrid model, you can own part of the computing infrastructure, to cater the needs of average load of traffic, 
and be able to move on to cloud computing whenever there is a spike in traffic/demand. 
You should also consider the network costs.
Because if it is a hybrid model,
you need a good network that connects your private infrastructure to the public cloud infrastructure. 
Costs for interoperability between private cloud and public clouds should be considered. 

So to summarize, utility pricing is good when demand varies over time.
As is the case for startup companies, seasonal businesses, etc. 
When utility premium is less than the ratio of peak demands to average demand, cloud computing is beneficial and makes sense.
