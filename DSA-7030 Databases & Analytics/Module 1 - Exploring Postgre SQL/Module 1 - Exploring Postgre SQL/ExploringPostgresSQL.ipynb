First, let's connect to the PostgreSQL DBMS.

To open the terminal, follow the steps at: [JupyterHub terminal session](../resources/JupyterHub_Terminal.ipynb)

Now that we have opened a terminal, let's look at connecting to a PostgreSQL database.

To connect to the database type (paste) the following command into your terminal session.
The **dsa_ro_user** user is a shared user, with read-only permission to a number of databases and schemas within the database.

```Bash
psql -U dsa_ro_user -h pgsql dsa_ro
```

When asked for a password enter: **readonly**  
Note: The password nor any characters will echo (written on to the screen) during command line password entry.

This should result in a prompt, similar to the following:

```Bash
[scottgs@indigo ~]$ psql -U dsa_ro_user -h pgsql dsa_ro
Password for user dsa_ro_user: 
psql (11.3, server 10.5)
Type "help" for help.

dsa_ro=>
```


[See the `psql` command documentation](https://www.postgresql.org/docs/11/index.html)

## Viewing Tables

To view all of the tables in a database we can use the following command:

```
\dt
```



We can also view the specific table in PostgreSQL using the following command:

```
\d util_us_states
```

## Viewing Data

To view the all the data from the tableName table within out database run the following command.

```SQL
SELECT * FROM util_us_states;
```



The table contains both rows and columns. 
A row is the entire set of data for each individual entry into the table. 
A column is a specific attribute of a row.


[See the `SELECT` command documentation](https://www.tutorialspoint.com/postgresql/postgresql_select_query.htm)  
**This is the link to the SELECT statement syntax for Postgres.**

### Now Let's Try Viewing Portions of the Data

What if we only wanted to view thes state with the state_alpha_code for Missouri.

This will require using a condition on the SELECT statement.

```SQL
SELECT * FROM util_us_states WHERE state_alpha_code = 'MO';
```

The WHERE condition allows you to limit the returns from the statement based on the condition you added to it.

[See the `WHERE` clause documentation](http://www.tutorialspoint.com/postgresql/postgresql_where_clause.htm)  
**This is the link to the WHERE clause syntax for Postgres.**

Now let's have you try and write a query. 

What if I wanted the name all the data for Virginia?

```SQL
SELECT * FROM util_us_states WHERE state_name = 'VIRGINIA';
```


Now what if we wanted to see the data for the state with the `state_number_code` that is 35?

```SQL
SELECT * FROM util_us_states WHERE state_number_code = 35;
```
